```python
reviews.columns = (
    reviews.columns
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9]+', '_', regex = True))
```

```python
#converting text blanks to NaN
text_cols = ["review_text","review_title","skin_tone","eye_color","hair_color","skin_type","product_name","brand_name"]
for col in text_cols:
    if col in reviews.columns:
        reviews[col] = reviews[col].replace("", np.nan)
```

```python
#numeric
num_cols = ["rating", "price_usd", "total_feedback_count", "total_pos_feedback_count", "total_neg_feedback_count", "helpfulness"]
for col in num_cols:
    if col in reviews.columns:
        reviews[col] = pd.to_numeric(reviews[col], errors = "coerce")
```

```python
#datetime
reviews["submission_date"] = pd.to_datetime(reviews["submission_date"], errors = "coerce")
```

```python
#string cleaning
string_cols = ["brand_name","product_name","skin_type","review_text","review_title"]
for col in string_cols:
    if col in reviews.columns:
        reviews[col] = reviews[col].astype("string").str.strip()
```

```python
#drop rows with no product or brand
reviews = reviews.dropna(subset = ["brand_name","product_name"])
```

```python
#impute missing rating with average rating
reviews["rating"] = reviews["rating"].fillna(reviews["rating"].mean())
```

```python
#impute missing price with median price
reviews["price_usd"] = reviews["price_usd"].fillna(reviews["price_usd"].median())
```

```python
#converting recommendation to boolean
reviews["is_recommended"] = reviews["is_recommended"].astype(str).str.lower().map({
    "true": True, "false": False, "yes": True, "no": False, "1": True, "0": False})
```

```python
#price tier columns
bins = [-np.inf, 25, 50, np.inf]
labels = ["$", "$$", "$$$"]
```

```python
reviews["price_tier"] = pd.cut(reviews["price_usd"], bins = bins, labels = labels)
reviews["price_tier"].value_counts()
```

```python
#adding word count of review text
reviews["review_length_words"] = reviews["review_text"].str.split().str.len()
```

```python
#skin types
reviews["skin_type"] = reviews["skin_type"].str.lower()
```

```python
#map other similar categories
reviews["skin_type_clean"] = reviews["skin_type"].replace({
    "combo":"combination",
    "balanced":"normal"})
```

```python
for c in ["review_text","review_title","skin_tone","eye_color","hair_color","skin_type","product_name","brand_name"]:
    if c in reviews.columns:
        reviews[c] = reviews[c].astype("string").str.strip()
        reviews.loc[reviews[c].str.len() == 0, c] = pd.NA
```

```python
must_have = [c for c in ["brand_name","product_name"] if c in reviews.columns]
if must_have:
    reviews = reviews.dropna(subset = must_have)
```

```python
if {"price_usd","brand_name"}.issubset(reviews.columns):
    brand_med = reviews.groupby("brand_name")["price_usd"].transform("median")
    reviews["price_usd"] = reviews["price_usd"].fillna(brand_med).fillna(reviews["price_usd"].median())
```

```python
#fixing my mistakes from Day 2 : changing columns to categorical
def norm_text(series):
    return (series.astype("string")
            .str.strip()
            .str.replace(r"\s+", " ", regex = True)
            .str.lower())
```

```python
if "eye_color" in reviews.columns:
    eye = norm_text(reviews["eye_color"]).replace({"grey":"gray"})
    eye = eye.fillna("unknown")
    eye_cats = ["brown", "blue", "hazel", "green", "gray", "unknown"]
    reviews["eye_color"] = pd.Categorical(eye, categories = eye_cats, ordered = False)
```

```python
if "hair_color" in reviews.columns:
    hair = norm_text(reviews["hair_color"]).fillna("unknown")
    hair_cats = ["black", "brown", "blonde", "red", "gray", "other", "unknown"]
    hair = hair.where(hair.isin(hair_cats), "other")
    reviews["hair_color"] = pd.Categorical(hair, categories = hair_cats, ordered = False)
```

```python
if "skin_tone" in reviews.columns:
    tone = norm_text(reviews["skin_tone"]).fillna("unknown")
    tone = tone.replace({"very fair":"fair", "light medium":"light", "medium deep":"deep", "dark deep":"deep", "deep dark":"deep"})
    tone_cats = ["fair", "light", "medium", "tan", "deep", "rich", "unknown"]
    tone = tone.where(tone.isin(tone_cats), "unknown")
    reviews["skin_tone"] = pd.Categorical(tone, categories = tone_cats, ordered = True)
```

```python
if "skin_type" in reviews.columns:
    st = norm_text(reviews["skin_type"]).replace({"combo":"combination", "balanced":"normal"})
    st = st.fillna("unknown")
    st_cats = ["dry", "normal", "oily", "combination", "unknown"]
    st = st.where(st.isin(st_cats), "unknown")
    reviews["skin_type"] = pd.Categorical(st, categories = st_cats, ordered = False)
```

```python
#I think I want to drop all of the unknowns since they won't be useful to me for my results
cats = ["eye_color", "hair_color", "skin_tone", "skin_type"]
```

```python
for c in cats:
    if c in reviews.columns:
        if reviews[c].dtype.name != "category":
            reviews[c] = reviews[c].astype("string")
        reviews = reviews.loc[reviews[c] != "unknown"].copy()
```

```python
for c in cats:
    if c in reviews.columns and reviews[c].dtype.name == "category":
        reviews[c] = reviews[c].cat.remove_categories(["unknown"]).cat.remove_unused_categories()
```

```python
#processing strings
reviews["brand_name_clean"] = reviews["brand_name"].astype("string").str.strip().str.lower().str.replace(r"\s+"," ", regex = True)
reviews["product_name_clean"] = reviews["product_name"].astype("string").str.strip().str.lower().str.replace(r"\s+"," ", regex = True)
reviews["review_title_clean"] = reviews["review_title"].astype("string").str.strip()
reviews["review_text_clean"] = reviews["review_text"].astype("string").str.strip()
```

```python
import re

reviews["review_title_alpha"] = reviews["review_title_clean"].fillna("").apply(lambda x: re.sub(r"[^A-Za-z0-9\s]", "", x)).str.strip()
reviews["review_title_tcase"] = reviews["review_title_alpha"].str.title()
```

```python
kw_pos = r"\b(amazing|love|great|excellent|holy grail|works well)\b"
kw_neg = r"\b(bad|hate|irritate|burn|break\s?out|waste)\b"

reviews["has_pos_kw"] = reviews["review_text_clean"].str.contains(kw_pos, case = False, na = False)
reviews["has_neg_kw"] = reviews["review_text_clean"].str.contains(kw_neg, case = False, na = False)
```

```python
reviews["review_len_words"] = reviews["review_text_clean"].str.split().str.len()

reviews["review_nuniq_words"] = (
    reviews["review_text_clean"]
      .fillna("")
      .str.lower()
      .str.replace(r"[^a-z0-9\s]", " ", regex = True)
      .str.split()
      .apply(lambda toks: len(set(toks)))
)
```

```python
reviews["product_name_clean"] = (
    reviews["product_name"]
      .astype("string")
      .str.strip()
      .str.lower()
      .str.replace(r"\s+", " ", regex = True))
```

```python
mini_patterns = r"(?:\bmini\b|\bmini[-\s]?size\b|\btravel[-\s]?size\b|\bdeluxe\s+sample\b|\bminiature\b)"

reviews["size_label"] = np.where(
    reviews["product_name_clean"].str.contains(mini_patterns, case = False, na = False),
    "mini",
    "full")
```

```python
reviews["size_label"] = pd.Categorical(reviews["size_label"], categories = ["mini","full"], ordered = True)
```

```python
reviews["spf"] = (
    reviews["product_name_clean"]
      .str.extract(r"\bspf\s*([0-9]{1,3})\b", expand = False)
      .astype(float))
```

```python
reviews["is_fragrance_free"] = reviews["review_text_clean"].str.contains(r"fragrance[-\s]?free", case = False, na = False)
reviews["is_alcohol_free"] = reviews["review_text_clean"].str.contains(r"alcohol[-\s]?free", case = False, na = False)
```

```python
reviews["brand_product_key"] = reviews["brand_name_clean"].fillna("") + " :: " + reviews["product_name_clean"].fillna("")
```

```python
split2 = reviews["product_name_clean"].str.split(r",|\bwith\b|\band\b", n = 1, regex = True, expand = True)
reviews["prod_lead"] = split2[0].str.strip()
reviews["prod_desc"] = split2[1].str.strip()
```

```python
product_type_rules = {
    "cleanser": ["cleanser", "face wash", "gel cleanser", "foam cleanser"],
    "moisturizer": ["moisturizer", "cream", "hydrator", "gel cream", "lotion"],
    "serum": ["serum", "essence", "ampoule"],
    "mask": ["mask", "face mask", "sheet mask", "clay mask"],
    "toner": ["toner", "astringent"],
    "sunscreen": ["sunscreen", "spf", "sun cream", "sun block"],
    "foundation": ["foundation", "tint", "bb cream", "cc cream"],
    "primer": ["primer"],
    "concealer": ["concealer", "corrector"],
    "setting spray": ["setting spray", "fix mist"],
    "lip product": ["lipstick", "lip gloss", "lip balm", "lip oil", "lip tint"],
    "eye product": ["eyeliner", "mascara", "eyeshadow", "brow", "lash"],
    "fragrance": ["perfume", "eau de", "fragrance"],
    "hair product": ["shampoo", "conditioner", "hair oil", "hair mask"],
    "body product": ["body lotion", "body wash", "body scrub"]}
```

```python
def classify_product(name):
    for ptype, keywords in product_type_rules.items():
        for kw in keywords:
            if kw in name:
                return ptype
    return "other"

reviews["product_type"] = reviews["product_name_clean"].apply(classify_product)
reviews["product_type"] = reviews["product_type"].astype("category")
```

```python
#one way anova : do skin types give different rating?
from scipy.stats import f_oneway
```

```python
anova_data = [
    reviews.loc[reviews["skin_type"] == st, "rating"].dropna()
    for st in reviews["skin_type"].unique()]
```

```python
f_stat, p_value = f_oneway(* anova_data)
print("F-statistic:", f_stat)
print("p-value:", p_value)

#since p-value < 0.05, skin type has a statistically significant effect on product ratings
```

```python
#two sample t-test : mini vs full size ratings
from scipy.stats import ttest_ind
```

```python
mini = reviews.loc[reviews["size_label"] == "mini", "rating"].dropna()
full = reviews.loc[reviews["size_label"] == "full", "rating"].dropna()
```

```python
t_stat, p_value = ttest_ind(mini, full, equal_var = False)
print("T-statistic:", t_stat)
print("p-value:", p_value)

#since p-value > 0.05, size does not sigificantly impact rating
```

```python
#chi-square test : is recommendation rating different by skin type?
import scipy.stats as stats
```

```python
ct = pd.crosstab(reviews["skin_type"], reviews["is_recommended"])
chi2, p, dof, expected = stats.chi2_contingency(ct)
```

```python
print("Chi-square:", chi2)
print("p-value:", p)
print("\nExpected counts:\n", expected)

#since p-value > 0.05, no evidence that skin types recommend products differently
```
